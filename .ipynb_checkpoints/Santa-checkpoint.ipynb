```python
from random import shuffle
```

```python
peeps = {1:"Bryan", 2: "Will", 3: "Lee", 4: "Eric", 5: "Michelle", 6: "Vincent", 7: "Brooke", 8: "Annie", 9: "Riana"}
# print len(peeps)
l = [i+1 for i in range(len(peeps))]
l2 = [i+1 for i in range(len(peeps))]
shuffle(l2)
dissallow_rules = [(4,5), (5,4), (6,7), (7,6), (8,2), (2,8),(1,9),(9,1)]
force_rules = [(2,1)]
```

```python
# print zip(l, l2)
print "dissallow_rules:", dissallow_rules
print "force_rules:", force_rules
soln = False
```

```python
while not soln:
    soln = True
    shuffle(l2)
    matches = zip(l,l2)
    print matches
    for p in matches:
        if p in dissallow_rules or p[0]==p[1]:
            print p
            soln = False
    if soln:
        for r in force_rules:
            if r not in matches:
                print r
                soln = False
    print soln
```
